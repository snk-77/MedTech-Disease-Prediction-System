```python
import pandas as pd
# Load the training dataset
train_data = pd.read_csv("/content/Training.csv.zip")
# Load the testing dataset
test_data = pd.read_csv("/content/Testing.csv")
```

 Preprocessing

```python
train_data.drop(["Unnamed: 133"], axis=1, inplace=True)
train_data.columns
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for column in train_data.columns:
    if train_data[column].dtype == 'object':  # Check if the column is categorical
        train_data[column] = le.fit_transform(train_data[column])

# Similarly, for the test dataset:
for column in test_data.columns:
    if test_data[column].dtype == 'object':
        test_data[column] = le.fit_transform(test_data[column])
```

```python
train_columns = train_data.columns
test_columns = test_data.columns
```

```python
for column in train_columns:
    train_data[column].fillna(train_data[column].median(), inplace=True)

for column in test_columns:
    test_data[column].fillna(test_data[column].median(), inplace=True)
```

```python
train_data.isnull().sum()
```

frequency analysis

```python
import matplotlib.pyplot as plt

# Select the diseases to visualize
selected_diseases = train_data['prognosis'].unique()[:5]  # Adjust as needed

# Create a figure and axes
fig, ax = plt.subplots(figsize=(12, 8))

# Initialize a DataFrame to store symptom counts
symptom_counts_df = pd.DataFrame()

# Iterate through selected diseases
for disease in selected_diseases:
    subset = train_data[train_data['prognosis'] == disease]
    symptom_counts = subset.drop('prognosis', axis=1).sum()
    symptom_counts_df[disease] = symptom_counts

# Select top 10 symptoms and remove zero counts
top_symptoms = symptom_counts_df.sum(axis=1).sort_values(ascending=False).head(10).index
symptom_counts_df = symptom_counts_df.loc[top_symptoms]

# Plot the stacked bar chart
symptom_counts_df.plot(kind='bar', stacked=True, ax=ax)

# Set labels and title
ax.set_xlabel('Symptom')
ax.set_ylabel('Count')
ax.set_title('Frequency of Top 10 Symptoms for Selected Diseases (Stacked)')
ax.legend()
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Test and train model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train = train_data.drop('prognosis', axis=1)
y_train = train_data['prognosis']
X_test = test_data.drop('prognosis', axis=1)
y_test = test_data['prognosis']

# Split training data for validation
X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.2)

# Create and train the model
model = RandomForestClassifier()
model.fit(X_train_split, y_train_split)

```

 Feature Importance

```python
feature_importances = pd.Series(model.feature_importances_, index=X_train.columns)
feature_importances.nlargest(10).plot(kind='barh')
plt.title('Top 10 Feature Importances')
plt.xlabel('Importance Score')
plt.ylabel('Symptom')
plt.show()
```

Accuracy of the model

```python
# Evaluate on validation set
y_val_pred = model.predict(X_val)
val_accuracy = accuracy_score(y_val, y_val_pred)
print("Validation Accuracy:", val_accuracy)

# Test on the test set
y_test_pred = model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_test_pred)
print("Test Accuracy:", test_accuracy)

# test f1 score
from sklearn.metrics import f1_score

# Evaluate on validation set
y_val_pred = model.predict(X_val)
val_f1 = f1_score(y_val, y_val_pred, average='weighted')
print("Validation F1-Score:", val_f1)

# Test on the test set
y_test_pred = model.predict(X_test)
test_f1 = f1_score(y_test, y_test_pred, average='weighted')
print("Test F1-Score:", test_f1)
```
